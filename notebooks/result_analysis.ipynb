```python
import pandas as pd
from datetime import datetime, timedelta, date
import mplfinance as mpf
import pandas_ta as pta
import numpy as np
import csv, os


def agg_results(x): 
    agg_values = {
        "wins": x[x["is_win"]==True]["is_win"].count(),
        "losses": x[x["is_win"]==False]["is_win"].count()
    }
    agg_values["total_trades"] = (agg_values["losses"] + agg_values["wins"])
    agg_values["win_pcnt"] = (agg_values["wins"] / agg_values["total_trades"]) * 100
    agg_values["profit"] = x["profit"].sum()
    return pd.Series(agg_values)


data = pd.read_csv("../logs/trade_results.csv")
data["trade_time"] = data["time"]
data["ftime"] = pd.to_datetime(data["time"]).dt.strftime("%I:%M %p")
data.set_index("time", inplace=True)

# create stats
data["is_win"] = np.where(data.profit >= 0, True, False)
data["month"] = pd.to_datetime(data["trade_time"]).dt.month.astype(str).str.pad(2, fillchar="0")
data["year"] = pd.to_datetime(data["trade_time"]).dt.year.astype(str)
data["week"] = pd.to_datetime(data["trade_time"]).dt.isocalendar().week.astype(str).str.pad(2, fillchar="0")

data["year_month"] = data["year"] + "-" + data["month"]
data["year_week"] = data["year"] + "-" + data["week"] 



```

```python

#result = data.groupby(["year_month","symbol","strategy", "timeframe"]).apply(agg_results)
#result = data.groupby([ "symbol", "year_week"]).apply(agg_results)
result = data.groupby(["year_week","strategy"]).apply(agg_results)

result
```
